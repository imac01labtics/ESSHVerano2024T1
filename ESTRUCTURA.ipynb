### Importaciones

```python
import CodeBase.Datos_Fa_Fr_FrAc as simpleData # Contiene funciona para, clases y fa, fr, f acumulada o todas en una sola funcion 
import CodeBase.DatosDiscretosYCualitativos as dtDisc # Contiene una funcion que retorna clases, fa, fa acumumlada, fr, fr acumumlada 
import CodeBase.DatosContinuos as dtCont # Contiene funciona que retorna, clases, limites, marcas de Clases, fa, fr, fr acumumlada
import CodeBase.GenerarGraficas as graf # Contiene un método para cada tipo de gráfica o para llamar todas al mismo tiempo 
import CodeBase.CrearTabla as tb # Método para crear tabla
```

## Ejemplo de Frecuencia Absoluta, Frecuencia Relativa y Acumulada

```python
datos = [5,2,2,4,5,1,1,4,2,1,3,5,1,0,5,0,3,2,3,4]

# Obtenemos clases, fa, fr, f acumulada (Ya están ordenados)
clases, fa, fr, frAc = simpleData.calcularFaFrFrAc(datos)

# Agregar simbolo de porcentaje %
frStr, frAcStr = [], []
for i in range(len(fr)):
    frStr.append(str(fr[i]) + "%")
for i in range(len(frAc)):
    frAcStr.append(str(frAc[i]) + "%")
    
# Creamos la tabla
encabezados = ["Clases", "     Fa", "     Fr", "  Fr Acum."]
contenido = [clases, fa, frStr, frAcStr]
tb.printTable(encabezados, contenido)
```

## Ejemplo para datos Discretos *(solo números enteros)* o Cualitativos *(Cadenas)*

```python
########### DATOS DISCRETOS ###########
datosDiscretos = [
    327,429,470,550,340,250,650,326,530,250,302,280,206,645,350,956,360,1067,635,553,730,630,250,1000,240,
    1000,420,150,160,200,150,550,400,330,115
]

# Obtenemos clases, fa, fa acumulada, fr, fr acumulada (Ya están ordenados)
clases, fa, faAc, fr, frAc = dtDisc.generateDiscreteData(datosDiscretos)

# Agregar simbolo de porcentaje %
frStr, frAcStr = [], []
for i in range(len(fr)):
    frStr.append(str(fr[i]) + "%")
for i in range(len(frAc)):
    frAcStr.append(str(frAc[i]) + "%")

# Creamos la tabla
encabezados = ["Clases", "     Fa", "      Fr", "   Fr Acum."]
contenido = [clases, fa, frStr, frAcStr]
tb.printTable(encabezados, contenido)

# Creamos graficas
graf.createFrequencyPolygon(clases,fr)
graf.createOjiva(clases, frAc)
```

## Ejemplo para datos Continuos *(con punto decimal)* [En este hay 2 formas de realizarla]

```python
########### DATOS CUALITATIVOS ###########
datosContinuos = [0,0,0.33,0.33,0.33,0.33,0.66,0.66,0.66,0.66,0.99,0.99,0.99,1.32,1.32,1.32,1.65,1.65,1.65,1.65]

# Obtenemos todos los datos (Ya están ordenados)
# Primera forma
clases, limsInf, limsSup, mrksClases, fa, fr, frAc = dtCont.generateGroupedData(datos,1)
# Segunda forma
clases2, limsInf2, limsSup2, mrksClases2, fa2, fr2, frAc2 = dtCont.generateGroupedData(datos,2)

# Agregar simbolo de porcentaje
frStr, frAcStr, frStr2, frAcStr2 = [], [], [], []
for i in range(len(fr)):
    frStr.append(str(fr[i]) + "%")
    frStr2.append(str(fr2[i]) + "%")
for i in range(len(frAc)):
    frAcStr.append(str(frAc[i]) + "%")
    frAcStr2.append(str(frAc2[i]) + "%")
    
# Creamos la tabla de la primer ambas formas (SOLAMENTE OCUPAS UNA FORMA)
encabezados = ["Clases", "Límites inferiores", "Límites superiores", "Marca de clase", "     Fa", "     Fr", "  FrAcum."]
contenido = [clases, limsInf, limsSup, mrksClases, fa, frStr, frAcStr]
contenido2 = [clases2, limsInf2, limsSup2, mrksClases2, fa2, frStr2, frAcStr2]
tb.printTable(encabezados, contenido) # forma 1
print('\n')
tb.printTable(encabezados, contenido2) # forma 2

# Creamos graficas
graf.createHistogram(clases, fa) # forma 1
graf.createHistogram(clases, fa2) # forma 2
```
